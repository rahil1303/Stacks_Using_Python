## STACKS
### STACKS: 
It is a data structure that stores the item in Last In First Out Manner.
LIFO: Stands for Last In First Out
It means that whichever element come in last then it leaves first.

### When do we use Stacks?
We use stacks data structure whenever we need the functionality in our application that utilizes the last data first. Example: Plates stacks, Books 

Computer Term Examples:
1. Back button in Internet Browser
2. Undo Functon
3. Most recently played music
4. Most recently visited websites
5. Bracket closing in programming of Python
6. Push, Pop operation in Operating systems, Recursion Function

### Operations of Stacks:
1. Create Stack: CustomStack()
2. Push: customStack.push(1)
3. Pop: customStack.pop()
4. Peek: The top element will return the top of stack value but it will not remove the top element from the stack. CustomStack.peek()
5. isEmpty(): customStack.isEmpty(). If the stack is empty then it will return true whereas if the stack is not empty then it will return false.
6. isFull(): If the size of the stack has reached maximum capacity then it wil return true.
7. DeleteStack: customStack.delete(). To delete the entire stack.

```python
### Create a STACK using Lists
class Stack:
    def __init__(self):
        self.lists = []
    def __str__(self):
        values = self.lists.reverse()
        values = [str(x) for x in self.lists]
        return "\n".join(values)
### Time Complexity is O(n)
### .isEmpty() Function
    def isEmpty(self):
        if self.lists == []:
            return True
        else:
            return False
### Time Complexity is O(1)

### Push() Method:
    def push(self,values):
        self.lists.append(values)
        return "The element has been added succesfully"
### Time Complexity is O(n) or O(n2)

    
### Pop() Method:
    def pop(self):
        if self.isEmpty():
            return "There is not any element in the stack"
        else:
            return self.lists.pop()
### Time Complexity is O(1)

        
### Peek Method()
    def peek(self):
        if self.isEmpty():
            return "there is no element is the stack"
        else:
            return self.lists[len(self.lists)-1]

    ### Time Complexity is O(1)

### Delete Entire Stack:
    def delete(self):
        self.lists = None
        return True
        

### Test:
customStack = Stack()
print(customStack.isEmpty())
customStack.push(1)
customStack.push(2)
customStack.push(3)
customStack.push(4)
print(customStack)
print("POP")
customStack.pop()
print(customStack.pop())
print(customStack)
print("PEEK")
print(customStack)
print(customStack.peek())
print("Delete")
customStack.delete()
```
